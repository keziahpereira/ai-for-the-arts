# The Oracle of Forgotten Futures — Designing a Speculative AI Installation

## 1. Overview

**The Oracle of Forgotten Futures** is a poetic, text-based AI application that invites users to interact with a fictional Oracle—a large language model (LLM) trained on a curated set of decolonial, feminist, Indigenous, and speculative literature. The Oracle responds to open-ended, imaginative questions with symbolic, dreamlike text. Rather than giving answers or facts, it encourages creative thinking, cultural reflection, and emotional engagement.

The design centers on slowing down digital interaction, making space for voices and ideas that have been marginalized or forgotten, and reimagining AI not as a problem-solver, but as a thoughtful co-creator.

---

## 2. Interaction Design

### Who Interacts?

- **Human Users**: Individuals or small groups can ask questions through a simple text input (online or in an installation setting).
- **The Oracle (AI)**: A single large language model trained on a custom corpus of marginalized, speculative, and poetic texts.
- **Optional Curator or Guide**: In physical settings, a facilitator could help guide reflection and discussion.

### How the Interaction Works:

1. **User Input**: A visitor types a question (e.g., “What stories live in the shadows of lost cities?”).
2. **LLM Output**: The Oracle generates a speculative response in poetic prose (e.g., blending history, mythology, dream logic).
3. **Response Display**: The text appears slowly on-screen, accompanied by abstract visuals and ambient sound.
4. **User Reflection**: Visitors are encouraged to save, annotate, or contribute their own responses to a growing archive.

### Tools Used:

- LLM (e.g., GPT or fine-tuned LLaMA model)
- Simple front-end (text input + visual/sound output)
- Optional archive system (e.g., static website or shared document)

---

## 3. Value of the Design

### Learning

- Encourages deeper, emotional learning through speculation and creative questioning.
- Introduces users to underrepresented knowledge systems and voices.
- Makes abstract concepts (like borders, memory, futures) feel more personal and tangible.

### Engagement

- Turns interaction with AI into a reflective and artistic experience.
- Builds a collective archive of questions and responses that evolves over time.
- Encourages visitors to contribute, discuss, and return.

### Inclusion

- Prioritizes texts and voices often excluded from mainstream AI datasets.
- Offers an alternative model of AI—gentle, poetic, and speculative—rather than dominant productivity-focused uses.
- Makes space for users from diverse backgrounds to ask and imagine freely.

---

## 4. Critical Reflection

### Potential Positive Impacts

- **Cultural memory**: Supports remembering and imagining beyond dominant histories.
- **Creative thinking**: Promotes non-linear, poetic exploration with AI as a thought partner.
- **Emotional space**: Offers a quiet, non-judgmental space for wonder, grief, hope, and curiosity.

### Potential Risks

- **Misinterpretation**: Users might expect factual answers and misunderstand the Oracle’s speculative nature.
- **Bias in training**: Even a curated dataset can carry bias or incomplete representation.
- **Over-romanticizing AI**: There is a risk of giving too much symbolic power to the machine.

### Ethical Design Choices

- Make the speculative and artistic nature of the Oracle clear to users.
- Include a disclaimer about the limits of AI understanding.
- Regularly review and expand the dataset to be more inclusive and sensitive.

---

## 5. Conclusion

**The Oracle of Forgotten Futures** is an experiment in rethinking what AI can be—less about utility, and more about collective imagination. It doesn’t aim to give answers but to hold space for questions that might not yet have them. It invites users to listen differently, imagine deeply, and reflect on the stories and futures that mainstream systems have often ignored.

This project challenges the idea that AI always needs to be efficient or productive. Instead, it suggests that AI could also help us slow down, explore uncertainty, and connect with ideas that don’t usually get attention. It’s a reminder that not all knowledge fits into categories or answers.

By focusing on voices from decolonial, feminist, and Indigenous traditions, the Oracle becomes a quiet space where different perspectives are heard and valued. It’s not perfect, and it doesn’t pretend to be. But it opens a doorway to engage with AI more thoughtfully—as a companion in questioning, rather than a machine that claims to know it all.

In the future, designs like this could inspire other creative uses of AI—ones that support emotional learning, cultural reflection, and inclusive storytelling. The Oracle may not solve problems, but it offers something just as important: space to wonder, imagine, and dream together.
