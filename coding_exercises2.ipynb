# 6. Using the matplotlib library to look at a picture as numerical data
In this exercise, we explore how an image can be represented and manipulated as numerical data using the `matplotlib` and `numpy` libraries in Python.

```python
import matplotlib.pyplot as pyplot
import numpy as np
```

```python
from matplotlib import pyplot


test_picture = pyplot.imread("picture1.jpg")

print("Numpy array of the image is: ", test_picture)


pyplot.imshow(test_picture)
pyplot.axis('off')  
pyplot.title("Original Image") 
pyplot.show()
```

### Step 2: Understanding the image as numerical data

Each pixel in the image is stored as a set of numerical values (e.g., RGB values). The image is loaded into a Numpy array where:
- Each row = image height (Y)
- Each column = image width (X)
- Each "depth" level = R, G, B (and sometimes alpha)

The `print()` command helps us view this underlying array.


```python

test_picture_filtered = 2 * test_picture / 3


pyplot.imshow(test_picture_filtered)
pyplot.axis('on')
pyplot.title("Filtered Image (Brightness Reduced)")
pyplot.show()
```

### What do you see displayed? What is going on here?

After applying the filter to the image using the formula `test_picture_filtered = 2 * test_picture / 3`, the image appears much lighter and more abstract. The hedgehog in the original image is no longer clearly visible. Instead, we see many colourful dots and outlines where the features used to be.

I think this happened because we altered the pixel values that represent brightness and colour. By scaling them, we’ve reduced the intensity of the darker shades, making the contrast between different parts of the image weaker. 
This exercise helped me understand that images are essentially arrays of numbers, and even simple mathematical operations can dramatically change how they look.
